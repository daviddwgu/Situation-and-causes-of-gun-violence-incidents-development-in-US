```python
# import package
import pandas as pd
```

```python
# load data
data = pd.read_csv('data.csv', sep=',', error_bad_lines=False, index_col=False, dtype='unicode')
data
```

```python
# check the data information
data.info()
```

```python
# check the data ia all digit or not
for i in range(len(data)):
    if not data.loc[i,'n_injured'].isdigit():
        print(i)
        print(data.loc[i,'n_injured'])
```

```python
# extract the number of kill
kill = []
for i in list(data['n_killed']):
    kill.append(int(i))
kill
```

```python
# extract the number of unjured
injured = []
for i in list(data['n_injured']):
    injured.append(int(i))
injured
```

```python
# combine two data, and make column for number of killed or injured
data['n_injured'] = injured
data['n_killed'] = kill
data['killed_injured'] = data['n_killed'] + data['n_injured']
```

```python
# only extract the reqired columns
data = data.loc[0:(len(data)+1),['incident_id','date','state','killed_injured']]
data
```

```python
# extract the year data out of date
list_data = list(data['date'])
year_list = []
for i in list_data:
    year_list.append(int(i.split('/')[0]))
year_list
```

```python
# create column to record year data
data['year'] = year_list
```

```python
# check it
data
```

```python
# extract the state name
list_state = sorted(list(set(data['state'])))
list_state
```

```python
# create dataframe templete
temp = {'state':list_state,'2013':0,'2014':0, '2015':0, '2016':0, '2017':0,'2018':0, '2013_count':0, '2014_count':0, '2015_count':0, 
        '2016_count':0, '2017_count':0, '2018_count':0}
frame1 = pd.DataFrame(temp)
frame1
```

```python
# extract the data for each state
year = ['2013','2014','2015','2016','2017','2018']
year_count = ['2013_count','2014_count','2015_count','2016_count','2017_count','2018_count']
for i in range(len(frame1)):
    temp_data = data[data['state']==frame1.loc[i,'state']]
    for j in year:
        temp_temp_data = temp_data[temp_data['year']==int(j)]
        frame1.loc[i,j] = len(temp_temp_data)
        frame1.loc[i,year_count[year.index(j)]] = sum(list(temp_temp_data['killed_injured']))
frame1
```

```python
# the data of 2013 and 2018 are extrame low, it is not right, so drop it
frame1 = frame1.drop(['2013','2013_count'], axis=1)
frame1 = frame1.drop(['2018','2018_count'], axis=1)
frame1
```

```python
# make 2 sum up columns
frame1['kill'] = frame1['2014'] + frame1['2015'] + frame1['2016'] + frame1['2017']
frame1['count'] = frame1['2014_count'] + frame1['2015_count'] + frame1['2016_count'] + frame1['2017_count']
frame1
```

```python
# change the form of dataframe
frame2 = pd.DataFrame()
frame2
for i in range(len(frame1)):
    temp_data = frame1.iloc[i,]
    temp = {'state':temp_data['state'],'year':[2014,2015,2016,2017],'count': list(temp_data)[1:5], 'killed_injured':list(temp_data)[5:9]}
    temp_frame = pd.DataFrame(temp)
    frame2 = frame2.append(temp_frame)
frame2
```

```python
# reset the index
frame2 = frame2.reset_index(drop=True)
frame2
```

```python
# load the unemployment rate data
emp = pd.read_excel('emp-unemployment.xls')
emp
```

```python
#drop the noise row
emp = emp.drop([0,1,2,3,61,62,63,64,65], axis=0)
emp
```

```python
# drop more noise row
emp = emp.drop([57,58,59,60], axis=0)
emp
```

```python
#drop the first column
emp = emp.drop([list(emp.columns)[0]], axis=1)
emp
```

```python
# rename the columns
emp.columns = emp.iloc[0,]
emp
```

```python
# drop the columns name row
emp = emp.drop([4], axis=0)
emp
```

```python
# reset the index
emp = emp.reset_index(drop=True)
emp
```

```python
# extract the reqired columns
emp = emp.loc[0:len(emp),['Area', 2014, 2015, 2016, 2017]]
emp
```

```python
# rename the area name for the US
emp.loc[0,'Area'] = 'All'
emp
```

```python
# check if the state name in main dataframe did not have data in the unemployment rate dataframe
list_state_gun = sorted(list(set(frame2['state'])))
list_state_emp = sorted(list(set(emp['Area'])))
lost_state = []
for i in list_state_gun:
    if i not in list_state_emp:
        print('yes')
        lost_state.append(i)
lost_state
```

```python
# extract the unemployment data to main dataframe
frame2['unemployment'] = 0
for i in range(len(frame2)):
    temp_state = frame2.loc[i,'state']
    temp_year = frame2.loc[i,'year']
    frame2.loc[i,'unemployment'] = float(emp[emp['Area']==temp_state][temp_year])
frame2
```

```python
# output it
frame2.to_csv('q2_2.csv',index=False)
```
